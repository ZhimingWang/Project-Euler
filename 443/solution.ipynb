We immediately realize $g(n) - g(n-1) > 1$ iff $g(n) = 3n$ for all $n > 6$.

Proof: Let $g(n_k) = 3n_k$, let $\delta$ be the smallest positive integer s.t. $g(n_k + \delta) - g(n_k + \delta - 1) > 1$. Then $g(n_k + \delta - 1) = 3n_k + \delta - 1$, so $\gcd(n_k + \delta, 3n_k + \delta - 1) > 1$. Let it be $p$, then $p \mid 2n_k - 1$, and $\delta \equiv -n_k \equiv -1/2 \equiv (p-1)/2 \pmod p$. Due to the minimality of $\delta$, apparently $p$ is the smallest prime factor of $2n_k - 1$, and $\delta = (p-1)/2$. Then $g(n_k + \delta) = 3n_k + \delta - 1 + p = 3(n_k + \delta)$.

Therefore,

$$n_1 = 9,\ n_{k+1} = n_k + \frac{\mathrm{lpf}(2n_k - 1) - 1}{2}$$

where $\mathrm{lpf}$ is the least prime factor function, and for $n_k \le n < n_{k+1}$,

$$g(n) = 3n_k + (n - n_k) = 2n_k + n.$$
