Not hard to show

$$M(n) = (n+1)^2 - 1.$$

Then it boils down to solving the equation

$$(n+1)^2 - 1 = \frac{m}{m+1}{2},$$

which is the generalized Pell equation

$$(2m+1)^2 - 8(n+1)^2 = -7.$$

All solutions $(x,\, y) = (2m+1,\, n+1)$ are generated by $(1+\sqrt{8})(3+\sqrt{8})^k$ and $(-1+\sqrt{8})(3+\sqrt{8})^k$.

Mathematica is a safe bet for a calculation like this.
