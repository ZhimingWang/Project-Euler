This is actually the first time I looked at a Project Euler problem soon after it came out. I didn't bother to start early, and it took me a few hours between reading the problem description and submitting the result, mainly because my "algorithm" was run in two sittings and took *two hours*, but I still managed to squeeze in at 59th, landing me two badges. All in all, not bad.

My "algorithm" is quite embarrassing. First I wrote a brute force Python implementation with a little bit of caching to calculate the exact $g(x)$ sequentially.

Having noticed that the $y$'s are quite small, I decided to narrow down on the $y$s I tried, which was quickly cut down to just 2. Curiously, when we limit to 2, we're still mostly there, except we need to backtrack in a sequence of $g(x)$ where $g(x) + x$ stays constant to find the true maximum:

```
2 2: 2
5 2: 29
45 2: 73
101 2: 141
213 2: 149 156 7
304 2: 168 174 6
329 2: 193 206 13
431 2: 271 281 10
669 2: 545 553 8
1003 2: 1031 1041 10
2293 2: 1097
2829 2: 1749
5409 2: 1953
6468 2: 3616
9372 2: 8762 8824 62
27597 2: 9794
```

This property could probably be proven theoretically using the converging property of orbits.

Anyway, I simply brute forced all $l(x, 2)$ (which took ~3000s), then backtracked to brute force $g(x)$ until the sequence breaks (which took ~4300s to complete but the result was obvious earlier).

Caching could be added to vastly speed up the computations. I didn't because (1) initially it was unclear if it would take too much memory; (2) once it was clear the program would take on the order of at most a several hours to produce the result, I didn't bother to complicate the code.

In hindsight even this same embarrasing "algorithm" could be vastly speed up with some huristic n-ary search without implementing caching (and hence adding no memory footprint at all).
