Apparently the first $N = 10^6$ numbers with at least $N$ prime factors will have a common prefix $2^{N-l}$ where $l$ is small. We need to find $l$. Note that for each $l$ we simply need to generate all numbers between $2^l$ and $\frac{3^{l+1}}{2}$ with at least $l$ prime factors. To do this we can simply use a priority queue starting with one element $2^l$, multiplying by a prime $\le \frac{3^{l+1}}{2^l}$ or replacing a 2 with said prime each time. We start from $l=1$ and test each consecutive $l$ until we find one where $N = 10^6$ numbers can be generated by the method above. The entire process takes <1s.
