Looking at the delta matrix, the pattern is rather obvious:

$$
S(m,\,n)=
\begin{cases}
6m + 2n - 13, & m > n,\, m \ge 3,\\
8m - 11, & m = n \ge 2,\\
6n + 2m - 13, & n > m,\, n \ge 3,\\
1, & (m,\,n) = (1,\,2),\, (2,\,1),\\
0, & (m,\,n) = (0,\,0).
\end{cases}
$$

Next we need to find $(m,\, n)$ s.t. $S(m,\, n) = p^2$. Annoyingly enough the problem does not state $m,\, n \ge 2$ but actually has this restriction, so keep it in mind. $8m - 11 \equiv -3 \pmod8$ so it can't be a square; we only need to consider $6m + 2n - 13 = p^2$, where

$$2 \le n = \frac{p^2+13-6m}{2} \le m-1,$$

so
$$\frac{p^2+15}{8} \le m \le \frac{p^2+9}{6}.$$

Other than $p = 3$, which give one such pair $(3,\, 2)$, we have $$\frac{p^2+15}{8} \in \mathbb{N}$ and $\left\lfloor \frac{p^2+9}{6} \right\rfloor = \frac{p^2+5}{6}$, so we have

$$\frac{p^2+5}{6} - \frac{p^2+15}{8} + 1 = \frac{p^2-1}{24}$$

such $m$, each giving us an $(m,\, n)$ pair. Finally, $(m,\, n)$ can be reversed, so the total number is

$$2 + \sum_{p \ge 3}\frac{p^2-1}{12}.$$
