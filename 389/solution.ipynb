The distribution of sums of $k$ independent rolls of an $n$-sided dice is represented by the coefficients of

$$\left(\frac{\sum_{j=1}^n x^j}{n}\right)^k.$$

Denote $\frac{\sum_{j=1}^n x^j}{n}$ by $f_n(x)$, then given a distribution of number of rolls $\{p_k\}$, the distribution of sums is generated by

$$\sum_k p_k f_n(x)^k.$$

We can use this to calculate the distribution of $C$, then that of $O$, then that of $D$.

Eventually we want to calculate the variance of $I$ from the distribution of $D$. Denote $E(X)$ of a single 20-sided dice as $E_1$, and $E(X^2)$ as $E_2$. Then

$$E(I^2) = \sum_{j} p_j E\left[\left(\sum_{i=1}^j X_j\right)^2\right] = \sum_{j} p_j (j E_2 + j(j-1) E_1),$$
$$E(I) = \sum_{j} p_j E\left[\sum_{i=1}^j X_j\right] = \sum_{j} p_j \cdot jE_1,$$

so

$$\mathrm{Var}(I) = E_2 \sum j p_j + E_1^2 \left[\sum j(j-1)p_j - \left(\sum j p_j\right)^2\right].$$
