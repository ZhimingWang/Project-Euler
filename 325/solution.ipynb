Writing $y/x$ in continued fraction form, it's not hard to see that $(x,y)$ is a losing configuration iff the leading number of ones in the continued fraction is odd, i.e., $y/x$ is in the following forms with $k \ge 2$,

$$1 + \frac{1}{k},\ 1 + \frac{1}{1+\frac{1}{1+\frac{1}{k}}},\ 1 + \frac{1}{1+\frac{1}{1+\frac{1}{1+\frac{1}{1+\frac{1}{k}}}}},\ \dots$$

In other words, iff $y/x < [1;1,1,1,\dots] = \frac{\sqrt{5}+1}{2} = \phi$ the golden ratio.


Therefore,

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
\begin{aligned}
S(N)
&= \sum_{y=1}^N \sum_{x=\ceildphi{y}}^{y-1} x + y \\
&= \sum_{y=1}^N y(y - \floordphi{y} - 1) + \frac{1}{2}(y - \floordphi{y} - 1)(y + \floordphi{y}) \\
&= \sum_{y=1}^N (y - \floordphi{y} - 1) \cdot \frac{3y+\floordphi{y}}{2} \\
&= \sum_{y=1}^N (2y - \floortphi{y} - 1) \cdot \frac{2y+\floortphi{y}}{2} \\
&= \sum_{y=1}^N (2y^2 - y) - \sum_{y=1}^N \frac{\floortphi{y}^2 + \floortphi{y}}{2} \\
&= \frac{N(N+1)(4N-1)}{6} - \sum_{y=1}^N \frac{\floortphi{y}^2 + \floortphi{y}}{2}.
\end{aligned}
$$

Let

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
U(n) = \sum_{y=1}^N \frac{\floortphi{y}^2 + \floortphi{y}}{2},\
V(n) = \sum_{y=1}^N y \floortphi{y}.
$$

Also note that the following summations are always equivalent (easily verified):

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
\sum_{y=1}^{\floordphi{n}} \sum_{x=1}^{\floortphi{y}} f(x, y) =
\sum_{x=1}^{\floortphi{\floordphi{n}}} \sum_{y=\ceildphi{x}}^{\floordphi{n}} f(x, y) =
\sum_{x=1}^n \sum_{y=\ceildphi{x}}^{\floordphi{n}} f(x, y)
$$

for any expression $f(x,y)$.

Now we realize that

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
\begin{aligned}
U(\floordphi{n})
&= \sum_{y=1}^{\floordphi{n}} \sum_{x=1}^{\floortphi{y}} x \\
&= \sum_{x=1}^n x \sum_{y=\ceildphi{x}}^{\floordphi{n}} \\
&= \sum_{x=1}^n x (\floordphi{n} - \floordphi{x}) \\
&= \sum_{x=1}^n x (\floordphi{n} - \floortphi{x} + x) \\
&= \floordphi{n} \cdot \frac{n(n+1)}{2} + \frac{n(n+1)(2n+1)}{6} - V(n),
\end{aligned}
$$

so

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
V(n) = \floordphi{n} \cdot \frac{n(n+1)}{2} + \frac{n(n+1)(2n+1)}{6} - U(\floordphi{n}).
$$

On the other hand,

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
\begin{aligned}
V(\floordphi{n})
&= \sum_{y=1}^{\floordphi{n}} y \sum_{x=1}^{\floortphi{y}} \\
&= \sum_{x=1}^n \sum_{y=\ceildphi{x}}^{\floordphi{n}} y \\
&= \sum_{x=1}^n \frac{\floordphi{n}(\floordphi{n}+1)}{2} - \frac{\floordphi{x}(\floordphi{x}+1)}{2} \\
&= n \cdot \frac{\floordphi{n}(\floordphi{n}+1)}{2} - \sum_{x=1}^n \frac{(\floortphi{x}-x)(\floortphi{x}-x+1)}{2} \\
&= n \cdot \frac{\floordphi{n}(\floordphi{n}+1)}{2} - \sum_{x=1}^n \frac{x^2-x}{2} - \sum_{x=1}^n\frac{\floortphi{x}^2-\floortphi{x}}{2} + \sum_{x=1}^n x\floortphi{x} \\
&= n \cdot \frac{\floordphi{n}(\floordphi{n}+1)}{2} - \frac{(n-1)n(n+1)}{6} - U(n) + V(n).
\end{aligned}
$$

so

$$
\newcommand{\floor}[1]{\left\lfloor#1\right\rfloor}
\newcommand{\ceil}[1]{\left\lceil#1\right\rceil}
\newcommand{\floordphi}[1]{\floor{#1/\phi}}
\newcommand{\floortphi}[1]{\floor{\phi #1}}
\newcommand{\ceildphi}[1]{\ceil{#1/\phi}}
\newcommand{\ceiltphi}[1]{\ceil{\phi #1}}
U(n) = n \cdot \frac{\floordphi{n}(\floordphi{n}+1)}{2} - \frac{(n-1)n(n+1)}{6} + V(n) - V(\floordphi{n}).
$$

Therefore, we have obtained recursive formulae for $U(n)$ and $V(n)$. Recall that

$$S(N) = \frac{N(N+1)(4N-1)}{6} - U(N),$$

so we're done.

```python
#!/usr/bin/env python3

import functools
import math


phi = (1 + math.sqrt(5)) / 2


@functools.lru_cache(maxsize=None)
def U(n):
    if n == 0:
        return 0
    ndphi = math.floor(n / phi)
    return n * ndphi * (ndphi + 1) // 2 - (n - 1) * n * (n + 1) // 6 + V(n) - V(ndphi)


@functools.lru_cache(maxsize=None)
def V(n):
    if n == 0:
        return 0
    ndphi = math.floor(n / phi)
    return ndphi * n * (n + 1) // 2 + n * (n + 1) * (2 * n + 1) // 6 - U(ndphi)


def S(n):
    return n * (n + 1) * (4 * n - 1) // 6 - U(n)


def main():
    print(S(10 ** 16) % (7 ** 10))


if __name__ == "__main__":
    main()
```
